**Handling Mixed Text & Numeric Inputs in Time-Related Features**
time-related columns (Hours_slept, Work_hours, Hobby_hours, Number_of_tasks, and Social_interaction_hours) contained inconsistent formats. Participants entered values in different ways, including:
- Text descriptions (e.g., "I didn't do my hobbies today", "none")
- Ranges (e.g., "2-3", "5~6", "2 to 3")
- Sentences with multiple time values (e.g., "calling 1 hour, chatting 1 hour")
- Words instead of numbers (e.g., "Zero", "Alot")
- Missing values and symbols

Solution: A custom preprocessing function

- Convert plain text like "none" or "Zero" → 0
- Interpret "all day" / "most of the day" → 10 (representing high duration)
- Convert ranges like "2-3" or "2~3" → average (e.g., 2.5)
- Sum multiple values in a sentence (e.g., "1 hour, 1 hour" → 2)
- Extract and convert single valid numbers (e.g., "7.5" → 7.5)
- Leave missing values as NaN for later imputation

```python
#detect rows that have strings or any weird input other than a normal or float value.
import re

cols = ['Hours_slept', 'Work_hours', 'Hobby_hours',
        'Number_of_tasks', 'Social_interaction_hours']

def find_bad_numeric_rows(df, col):
    return df[
        # No number at all
        (df[col].astype(str).str.extract(r'(\d+\.?\d*)')[0].isna()) |

        # Has more than one numeric value BUT ignore decimals like 7.5
        (df[col].astype(str).apply(lambda x: len(re.findall(r'\d+(?:\.\d+)?', x)) > 1))
    ][[col]]

for c in cols:
    bad = find_bad_numeric_rows(Merged, c)
    if not bad.empty:
        print(f"\n⚠️ Suspicious values in '{c}':")
        print(bad)
```

```python
import re
import numpy as np

cols = ['Hours_slept', 'Work_hours', 'Hobby_hours',
        'Number_of_tasks', 'Social_interaction_hours']

# Function to convert messy text to a numeric value
def convert_to_numeric(val):
    if pd.isna(val):
        return np.nan

    s = str(val).strip().lower()

    # Hard mappings for textual values
    text_zero = ["none", "zero", "no", "didn't", "didnt", "no hobbies", "i didn't", "i didnt"]
    text_high = ["all day", "all day long", "most of the day"]

    if any(t in s for t in text_zero):
        return 0
    if any(t in s for t in text_high):
        return 10   # heavy interaction day

    # Extract numbers
    nums = re.findall(r'\d+(?:\.\d+)?', s)

    # If it contains multiple numbers like "calling 1 hour , chatting 1 hour" → sum = 2
    if len(nums) > 1:
        nums = list(map(float, nums))

        # If range like 2-3, 3~4 → average
        if re.search(r'(\d+)\s*[-~/]\s*(\d+)', s) or "to" in s:
            return sum(nums) / len(nums)

        # Else treat as multiple separate time segments → sum them
        return sum(nums)

    # One number only → return it
    if len(nums) == 1:
        return float(nums[0])

    # If no numbers exist and string isn't in zero/high lists, assume 0
    return 0

# Apply cleaner
for c in cols:
    Merged[c] = Merged[c].apply(convert_to_numeric)
```

**For the Stress_type column:**

The question was a checkbox that has 7 unique answers so, people could chose only one answer, or multiple ones, or not choose at all if they didn't experience any stressful event in their day.

The column's data is:
* Either comma or semicolon separated ( , /; )
* Null   , i.e There wasn't any stressful event.
* One choice  , i.e The cell has 1 normal string input.
* 2 or more choices , i.e The person has experienced multiple stressor in one day.

Solution:
* The Stress_type will be replace by other 7 columns representing the 7 choices.
* If the person didn't experience any stressful event(Nan), all the 7 columns will store zero.
* If only one choice got selected(Bad weather), the column presenting this choice will store 1 while the others store zeros.
* If more than one choice got chosen(,/; separated), All the columns of the chosen answers will have one and the others will have zeros.


```python
import pandas as pd

# Your stress type list
stress_types = [
    "Arguing/conflict with someone",
    "Academic/work failure or poor performance",
    "Transportation problem / car issue",
    "Personal disappointment",
    "Financial stress",
    "Health-related stress",
    "Bad weather"
]

# Create empty columns first
for st in stress_types:
    Merged[st] = 0

# Function to process each row
def process_stress_type(row):
    # If Stressful_event is 0 → all zeros
    if row['Stressful_event'] == 0 or pd.isna(row['Stress_type']):
        return pd.Series([0]*len(stress_types), index=stress_types)

    # Normalize separators & split
    values = str(row['Stress_type']).replace(';', ',').split(',')
    values = [v.strip() for v in values]  # clean spaces

    # Create 1/0 mapping
    out = {st: 0 for st in stress_types}
    for v in values:
        if v in out:
            out[v] = 1

    return pd.Series(out)

# Apply the function
Merged[stress_types] = Merged.apply(process_stress_type, axis=1)

# Drop the original column
Merged.drop(columns=['Stress_type'], inplace=True)
```

## Generate Synthetic Data using CTGAN (Tabular data augmentation)

### 1. Import CTGAN
Import the `CTGAN` model directly from the `ctgan` library. This is the correct import path after previous attempts with `sdv.tabular` and `sdv.single_table` failed.

```python
!pip install ctgan
from ctgan import CTGAN
```

### 2. Define Categorical and Continuous Columns
It's crucial to correctly identify categorical and continuous columns for CTGAN. Categorical columns are explicitly passed to the model, while continuous columns are inferred by default.


```python
categorical_cols = [
    'Age', 'Gender', 'Current_status', 'Task_difficulty',
    'Social_interaction_quality', 'Stressful_event', 'Home_environment',
    'Stress_scale', 'Overthinking', 'Stress_under_pressure',
    'Arguing/conflict with someone', 'Academic/work failure or poor performance',
    'Transportation problem / car issue', 'Personal disappointment', 'Financial stress',
    'Health-related stress', 'Bad weather'
]

continuous_cols = [
    'Hours_slept', 'Work_hours', 'Hobby_hours', 'Commute_time',
    'Number_of_tasks', 'Social_interaction_hours'
]

print("Categorical columns identified:", categorical_cols)
print("Continuous columns identified:", continuous_cols)
```

### 3. Initialize and Train CTGAN Model
Initialize the `CTGAN` model. We pass the identified `categorical_cols` to the `discrete_columns` parameter during the `fit` method. This trains the model to learn the patterns and distributions from your original `Merged` DataFrame.


```python
model = CTGAN()
# Train the CTGAN model with the Merged DataFrame, specifying discrete (categorical) columns
model.fit(Merged, discrete_columns=categorical_cols)
print("CTGAN model initialized and trained successfully.")
```

### 4. Generate Synthetic Data
Once the model is trained, we can use the `sample` method to generate the desired number of new synthetic rows (5000 in this case). These rows will statistically resemble the original data.


```python
# Generate 5000 new synthetic data rows
synthetic_data = model.sample(5000)
print("Generated 5000 synthetic rows.")

# Display the first few rows of the generated synthetic data
print("Displaying the first 5 rows of the generated synthetic data:")
synthetic_data.head()
```

```python
import numpy as np

# Function to convert Commute_time to specific numeric values
def convert_commute_time_synth(val):
    # Ensure value is treated as numeric, handling potential non-numeric outputs from CTGAN
    val = float(val)
    if val < 30:
        return 15
    elif 30 <= val < 60:
        return 45
    elif 60 <= val < 120: # Corrected logic to catch values between 60 and 120 and map to 60
        return 60
    elif val >= 120:
        return 150
    else:
        return 60 # Fallback to 60 for any unhandled edge cases (should not occur with above logic)

# Apply custom Commute_time logic to synthetic_data
synthetic_data['Commute_time'] = synthetic_data['Commute_time'].apply(convert_commute_time_synth)

# Columns to be float only, no negative values, and rounded to one decimal
float_cols = ['Hours_slept', 'Work_hours', 'Hobby_hours', 'Social_interaction_hours']
for col in float_cols:
    # Ensure no negative values (set to 0 if negative)
    synthetic_data[col] = np.maximum(0, synthetic_data[col])
    # Round to the first decimal place
    synthetic_data[col] = synthetic_data[col].round(1)

# Number_of_tasks: Should be integer only
synthetic_data['Number_of_tasks'] = synthetic_data['Number_of_tasks'].astype(int)

print("Preprocessing complete for specified columns in synthetic_data.")
print("Info after preprocessing:")
synthetic_data.info()
print("\nFirst 5 rows of synthetic_data after preprocessing:")
print(synthetic_data.head())
# Display unique values for Commute_time to confirm fix
print("\nUnique values for Commute_time after preprocessing:")
print(synthetic_data['Commute_time'].unique())
```
